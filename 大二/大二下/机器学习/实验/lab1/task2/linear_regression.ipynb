## Import packages

```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

## Load dataset and split

```python
data = fetch_california_housing(data_home='./data')
Xtrain, Xtest, ytrain, ytest = train_test_split(data.data, data.target, test_size=0.2, random_state=42)
```

## Create model, predict, evaluate and visualize

```python
linear_regressor = LinearRegression()
linear_regressor.fit(Xtrain, ytrain)
y_pred = linear_regressor.predict(Xtest)

mse = mean_squared_error(ytest, y_pred)
mae = mean_absolute_error(ytest, y_pred)
r2 = r2_score(ytest, y_pred)

print(f"MSE: {mse:.4f}")
print(f"MAE: {mae:.4f}")
print(f"RÂ²: {r2:.4f}")

plt.figure(figsize=(8, 6))
plt.scatter(ytest, y_pred, alpha=0.5, color='blue')
plt.plot([ytest.min(), ytest.max()], [ytest.min(), ytest.max()], 'k--', lw=2)  # çæ³å¯¹è§çº¿
plt.xlabel('True Values')
plt.ylabel('Predictions')
plt.title('True vs Predicted Values')
plt.show()
```
